```python
import numpy as np
import matplotlib.pyplot as plt 
import json
```

```python
path="results.json"
with open(path, "r") as f:
    results = json.load(f)
```

```python
paths=[]
for i in range (1,51):
    numbr="results"+str(i)+".json"
    paths.append(numbr)

result_dict={}
for p in paths:
    with open(p, "r") as f:
        result_dict[p] = json.load(f)
    
```

```python
#ANALYSIS FOR LEARNING RATE
```

```python
#Calculate the average for each changed learning rate

#e=0.1, lr=0.1, df=0.5, rewards=-10,10,1,-1
lr_1_avrg=[]
for ind in range (0,211):
    first_run=results[ind]['score']
    second_run=result_dict['results1.json'][ind]['score']
    third_run=result_dict['results2.json'][ind]['score']
    avrg=(first_run+second_run+third_run)/3
    lr_1_avrg.append(avrg)
    
#e=0.1, lr=0.3, df=0.5, rewards=-10,10,1,-1
lr_3_avrg=[]
for ind in range (0,211):
    first_run=result_dict['results3.json'][ind]['score']
    second_run=result_dict['results4.json'][ind]['score']
    third_run=result_dict['results5.json'][ind]['score']
    avrg=(first_run+second_run+third_run)/3
    lr_3_avrg.append(avrg)
    
#e=0.1, lr=0.5, df=0.5, rewards=-10,10,1,-1
lr_5_avrg=[]
for ind in range (0,211):
    first_run=result_dict['results6.json'][ind]['score']
    second_run=result_dict['results7.json'][ind]['score']
    third_run=result_dict['results8.json'][ind]['score']
    avrg=(first_run+second_run+third_run)/3
    lr_5_avrg.append(avrg)

#e=0.1, lr=0.7, df=0.5, rewards=-10,10,1,-1
lr_7_avrg=[]
for ind in range (0,211):
    first_run=result_dict['results9.json'][ind]['score']
    second_run=result_dict['results10.json'][ind]['score']
    third_run=result_dict['results11.json'][ind]['score']
    avrg=(first_run+second_run+third_run)/3
    lr_7_avrg.append(avrg)
    
#e=0.1, lr=0.9, df=0.5, rewards=-10,10,1,-1
lr_9_avrg=[]
for ind in range (0,211):
    first_run=result_dict['results12.json'][ind]['score']
    second_run=result_dict['results13.json'][ind]['score']
    third_run=result_dict['results14.json'][ind]['score']
    avrg=(first_run+second_run+third_run)/3
    lr_9_avrg.append(avrg)
```

```python
#plot
game_num=list(range(1,211+1))
plt.figure(figsize=(10,10))
plot1,=plt.plot(game_num, lr_1_avrg, 'gold')
plot2,=plt.plot(game_num, lr_3_avrg,'m')
plot3,=plt.plot(game_num, lr_5_avrg,'c')
plot4,=plt.plot(game_num, lr_7_avrg,'y')
plot5,=plt.plot(game_num, lr_9_avrg,'r')
plt.title('Learning rate analysis' )
plt.xlabel('Game number')
plt.ylabel('Score')
plt.legend([plot1,plot2,plot3,plot4,plot5],['Learning rate=0.1','Learning rate=0.3','Learning rate=0.5','Learning rate=0.7', 'Learning rate=0.9'])
```

```python
# line of best fit plot
plt.figure(figsize=(10,10))
m_1, b_1 = np.polyfit(game_num, lr_1_avrg, 1)
plot1,=plt.plot(game_num, m_1*np.array(game_num)+b_1, 'gold')
m_3, b_3 = np.polyfit(game_num, lr_3_avrg, 1)
plot2,=plt.plot(game_num, m_3*np.array(game_num)+b_3,'m')
m_5, b_5 = np.polyfit(game_num, lr_5_avrg, 1)
plot3,=plt.plot(game_num, m_5*np.array(game_num)+b_5,'c')
m_7, b_7 = np.polyfit(game_num, lr_7_avrg, 1)
plot4,=plt.plot(game_num, m_7*np.array(game_num)+b_7,'y')
m_9, b_9 = np.polyfit(game_num, lr_9_avrg, 1)
plot5,=plt.plot(game_num, m_9*np.array(game_num)+b_9,'r')
plt.title('Learning rate analysis line of best fit' )
plt.xlabel('Game number')
plt.ylabel('Score')
plt.legend([plot1,plot2,plot3,plot4,plot5],['Learning rate=0.1','Learning rate=0.3','Learning rate=0.5','Learning rate=0.7', 'Learning rate=0.9'])
```
